```python
import numpy as np
import gym
import time
 
env = gym.make('MountainCar-v0')
env.reset()
```

```python
def discrete_state(state , env):
  return np.round((state - env.observation_space.low)*np.array([10,100])).astype(int)
 
def train(env, learning_rate, discount_factor,number_of_episodes ,epsilon, dec_epsilon_rate):
    discrete_states = discrete_state(env.observation_space.high , env) + 1
    Q_table = np.zeros((discrete_states[0], discrete_states[1], env.action_space.n))
    
    for i in range(number_of_episodes):
        reward = 0
        current_state = discrete_state(env.reset() , env)
        end_episode = False
        while end_episode != True:   
            if np.random.random() < 1 - epsilon:
              action = np.argmax(Q_table[current_state[0], current_state[1]]) 
            else:
              action = np.random.randint(0, env.action_space.n)
 
            real_next_state, reward, end_episode, info = env.step(action) 
            Terminal = real_next_state[0] >= 0.5 and end_episode
            next_state = discrete_state(real_next_state , env)
 
            if Terminal:
                Q_table[current_state[0], current_state[1], action] = reward
                
            else:
                Q_table[current_state[0], current_state[1],action] += learning_rate*(reward + 
                          discount_factor*np.max(Q_table[next_state[0], next_state[1]]) - Q_table[current_state[0], current_state[1],action])
                                     
            current_state = next_state
            if (i+1)%100 == 0:
              epsilon = epsilon * dec_epsilon_rate
        
        if ((i+1) % 1000 == 0):
          print(str(i+1)+" episodes has trained")
    return Q_table
```

```python
start_time = time.time()
Q_table = train(env, 0.15, 1, 70000 , 1 , 0.9)
print("train time is :%s seconds " %(time.time() - start_time))
```

```python
# test
env = gym.make('MountainCar-v0')
current_state = discrete_state(env.reset(),env)
sum_of_rewards = 0
end_episode = False
while end_episode != True:
    action = np.argmax(Q_table[current_state[0], current_state[1]])
    next_state, reward, end_episode, info = env.step(action)
    next_state = discrete_state(next_state,env)
    sum_of_rewards += reward
    current_state = next_state
print(sum_of_rewards)
```

```python
res = []    
for i in range(5000):    
  env.close()
  env = gym.make('MountainCar-v0')
  current_state = discrete_state(env.reset(),env)
  sum_of_rewards = 0
  end_episode = False
  while end_episode != True:
    action = np.argmax(Q_table[current_state[0], current_state[1]])
    next_state, reward, end_episode, info = env.step(action)
    next_state = discrete_state(next_state,env)
    sum_of_rewards += reward
    current_state = next_state
  res.append(sum_of_rewards)
t1 = 0
t2 = 0
t3 = 0
for r in res:
  if r > -140:
    t1+=1
  elif r>-150:
    t2+=1
  else:
    t3+=1
 
print("in 5000 test")   
print("more than -140 : " + str(t1))
print("between -150 and -140 : "+ str(t2))
print("less than -150 : " + str(t3))
```
